```python
import pandas as pd 
```

```python
df=pd.read_excel("post_table.xlsx")
```

```python
df=df.dropna(subset=["post_id"])
```

```python
df=df.dropna(subset=["caption"])
```

```python
import numpy as np
from textblob import TextBlob
```

```python
hashtag_cols = [col for col in df.columns if col.startswith("hashtags")]
all_hashtags = []
for col in hashtag_cols:
    all_hashtags.extend(df[col].dropna().astype(str).tolist())


all_hashtags = [h.strip().lower() for h in all_hashtags if h not in ["", "nan", None]]

hashtag_freq = pd.Series(all_hashtags).value_counts().reset_index()
hashtag_freq.columns = ["hashtag", "frequency"]


def assign_category(tag):
    tag = tag.lower()

    # ---------- Travel & Nature ----------
    if any(word in tag for word in [
        "travel", "travelling", "trip", "vacation", "wander", "wanderlust",
        "adventure", "explore", "exploring", "journey", "roadtrip",
        "nature", "landscape", "mountains", "beach", "beaches",
        "sunset", "sunrise", "forest", "camping", "hiking",
        "trek", "backpacking", "outdoors", "wildlife", "waterfall",
        "hillstation", "nationalpark"
    ]):
        return "Travel & Nature"

    # ---------- Fitness & Health ----------
    elif any(word in tag for word in [
        "fitness", "fit", "gym", "workout", "training", "exercise",
        "health", "healthy", "healthylife", "wellness", "lifestyle",
        "yoga", "asana", "pranayama", "meditation",
        "running", "jogging", "homeworkout", "bodybuilding",
        "strength", "cardio", "crossfit", "fatloss", "weightloss"
    ]):
        return "Fitness & Health"

    # ---------- Fashion ----------
    elif any(word in tag for word in [
        "fashion", "style", "outfit", "ootd", "streetwear", "lookbook",
        "clothing", "apparel", "trendy", "trends",
        "mensfashion", "womensfashion", "ethnicwear", "casualwear",
        "formalwear", "summerfashion", "winterfashion",
        "accessories", "footwear", "shoes", "bags", "jewellery"
    ]):
        return "Fashion"

    # ---------- Food & Cooking ----------
    elif any(word in tag for word in [
        "food", "foodie", "foodporn", "cooking", "cook",
        "recipe", "recipes", "baking", "bake",
        "kitchen", "homemade", "homecooking",
        "healthyfood", "streetfood", "fastfood",
        "restaurant", "cafe", "coffee", "tea",
        "dessert", "cakes", "sweets", "snacks"
    ]):
        return "Cooking"

    # ---------- Technology & Programming ----------
    elif any(word in tag for word in [
        "tech", "technology", "programming", "coding", "developer",
        "software", "engineering", "computer", "computerscience",
        "python", "java", "javascript", "html", "css",
        "ai", "artificialintelligence", "machinelearning", "datascience",
        "cloud", "devops", "api", "backend", "frontend", "automation"
    ]):
        return "Technology & Programming"

    # ---------- Education ----------
    elif any(word in tag for word in [
        "study", "studying", "education", "learning", "learn",
        "student", "students", "college", "university",
        "books", "reading", "notes", "revision",
        "onlinecourses", "course", "courses",
        "science", "research", "exams", "testprep",
        "career", "skills", "knowledge"
    ]):
        return "Education"

    # ---------- Entertainment ----------
    elif any(word in tag for word in [
        "entertainment", "movies", "movie", "cinema",
        "webseries", "series", "tvshow",
        "comedy", "funny", "memes", "meme",
        "music", "song", "songs", "playlist",
        "dance", "reels", "viral", "trending",
        "celebrity", "actor", "actress", "film"
    ]):
        return "Entertainment"

    # ---------- Art & Creativity ----------
    elif any(word in tag for word in [
        "art", "artist", "creative", "creativity",
        "drawing", "sketch", "painting", "illustration",
        "digitalart", "graphicdesign", "design",
        "photography", "photo", "photoshoot",
        "animation", "motiongraphics",
        "3dart", "blender", "cgi", "visualart",
        "typography", "uiux", "posterdesign"
    ]):
        return "Art & Creativity"

    # ---------- Business & Finance ----------
    elif any(word in tag for word in [
        "business", "entrepreneur", "entrepreneurship", "startup",
        "company", "brand", "branding", "marketing",
        "finance", "finances", "money", "wealth",
        "investing", "investment", "stocks", "stockmarket",
        "trading", "crypto", "cryptocurrency",
        "ecommerce", "sales", "growth", "strategy"
    ]):
        return "Business & Finance"

    # ---------- Motivation & Self-Improvement ----------
    elif any(word in tag for word in [
        "motivation", "motivational", "inspiration", "inspirational",
        "selfhelp", "selfgrowth", "selfimprovement",
        "positivity", "mindset", "goals", "success",
        "discipline", "habits", "routine",
        "confidence", "believe", "focus",
        "mentalstrength", "personaldevelopment"
    ]):
        return "Motivation & Self-Improvement"

    # ---------- Sports ----------
    elif any(word in tag for word in [
        "sports", "sport", "cricket", "football", "soccer",
        "basketball", "tennis", "badminton",
        "athletics", "running", "cycling",
        "boxing", "mma", "wrestling",
        "esports", "match", "tournament",
        "league", "team", "stadium"
    ]):
        return "Sports"

    # ---------- Pets & Animals ----------
    elif any(word in tag for word in [
        "pets", "pet", "dog", "dogs", "cat", "cats",
        "animal", "animals", "petlover", "petlife",
        "puppy", "kitten",
        "petcare", "pettraining",
        "wildlife", "zoo", "birds",
        "aquatic", "fish", "vet"
    ]):
        return "Pets & Animals"

    # ---------- Relationships ----------
    elif any(word in tag for word in [
        "relationship", "relationships", "love", "couple",
        "couples", "dating", "marriage",
        "friendship", "friends", "bestfriends",
        "family", "parents", "siblings",
        "parenting", "kids", "children",
        "husband", "wife", "partner"
    ]):
        return "Relationships"

    # ---------- Beauty & Personal Care ----------
    elif any(word in tag for word in [
        "beauty", "skincare", "skin",
        "makeup", "makeuptutorial",
        "hair", "hairstyle", "hairstyles",
        "grooming", "nails", "nailart",
        "cosmetics", "selfcare",
        "facial", "spa", "wellgroomed",
        "personalcare"
    ]):
        return "Beauty & Personal Care"

    # ---------- Gaming ----------
    elif any(word in tag for word in [
        "gaming", "gamer", "games", "videogames",
        "pcgaming", "mobilegaming",
        "esports", "streaming",
        "livestream", "twitch",
        "pubg", "valorant", "csgo",
        "fortnite", "minecraft",
        "console", "playstation", "xbox"
    ]):
        return "Gaming"

    # ---------- Cars & Automobiles ----------
    elif any(word in tag for word in [
        "cars", "car", "automobile", "vehicles",
        "bikes", "bike", "motorcycle",
        "racing", "motorsport",
        "supercars", "luxurycars",
        "carmods", "carreview",
        "engine", "speed", "driving"
    ]):
        return "Cars & Automobiles"

    # ---------- Home & Interior ----------
    elif any(word in tag for word in [
        "home", "interior", "homedecor", "decor",
        "decoration", "furniture",
        "minimalism", "diy",
        "garden", "gardening",
        "plants", "houseplants",
        "architecture", "interiordesign",
        "livingroom", "bedroom"
    ]):
        return "Home & Interior"

    # ---------- News & Politics ----------
    elif any(word in tag for word in [
        "news", "breakingnews", "headlines",
        "politics", "political",
        "government", "election",
        "worldnews", "nationalnews",
        "currentaffairs",
        "policy", "debate", "opinion",
        "economy", "geopolitics"
    ]):
        return "News & Politics"

    # ---------- Default ----------
    else:
        return "Other"


hashtag_freq["category"] = hashtag_freq["hashtag"].apply(assign_category)


def get_sentiment(tag):
    analysis = TextBlob(tag).sentiment.polarity
    if analysis > 0:
        return "Positive"
    elif analysis < 0:
        return "Negative"
    else:
        return "Neutral"

hashtag_freq["sentiment"] = hashtag_freq["hashtag"].apply(get_sentiment)


hashtag_freq.to_csv("Hashtag_Table.csv", index=False)

print(hashtag_freq)
```
